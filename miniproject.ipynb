## Loading the dataset

```python
#importing
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the data set
sales1 = pd.read_csv('data/pc_sales_clean.csv')
```

## Subsetting the dataset

```python
#subsetting the dataset
year_2019 = sales1.loc[sales1.year == 2019]
year_2020 = sales1.loc[sales1.year == 2020]
year_2021 = sales1.loc[sales1.year == 2021]
```

```python
year_2019.head()
```

```python
year_2020.head()
```

```python
year_2021.head()
```

## Exploratory Data Analysis

```python
# plotting the data above

fig, ax = plt.subplots(figsize=(10,8))

x = year_2019['country']
y = year_2019['sales']

ax = sns.barplot(x = x, y = y)
plt.xticks(rotation = 90)

ax.set_title('Sales of 2019 Vs Country', fontsize=16)
ax.set_xlabel("Country",fontsize=14)
ax.set_ylabel("Sales", fontsize=14);
```

```python
# plotting the data above

fig, ax = plt.subplots(figsize=(10,8))

x = year_2020['country']
y = year_2020['sales']

ax = sns.barplot(x = x, y = y)
plt.xticks(rotation = 90)

ax.set_title('Sales of 2020 Vs Country', fontsize=16)
ax.set_xlabel("Country",fontsize=14)
ax.set_ylabel("Sales", fontsize=14);
```

```python
# plotting the data above

fig, ax = plt.subplots(figsize=(10,8))

x = year_2021['country']
y = year_2021['sales']

ax = sns.barplot(x = x, y = y)
plt.xticks(rotation = 90)

ax.set_title('Sales of 2019 Vs Country', fontsize=16)
ax.set_xlabel("Country",fontsize=14)
ax.set_ylabel("Sales", fontsize=14);
```
