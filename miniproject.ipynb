## Loading the dataset

```python
#importing
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the data set
sales1 = pd.read_csv('data/pc_sales_clean.csv')
```

## Subsetting the dataset

```python
#subsetting the dataset
year_2019 = sales1.loc[sales1.year == 2019]
year_2020 = sales1.loc[sales1.year == 2020]
year_2021 = sales1.loc[sales1.year == 2021]
```

```python
year_2019.head()
```

```python
year_2020.head()
```

```python
year_2021.head()
```

## Exploratory Data Analysis

### 1. Sales of 2019 vs country

```python
# plotting the data above

fig, ax = plt.subplots(figsize=(10,8))

x = year_2019['country']
y = year_2019['sales']

ax = sns.barplot(x = x, y = y)
plt.xticks(rotation = 90)

ax.set_title('Sales of 2019 Vs Country', fontsize=16)
ax.set_xlabel("Country",fontsize=14)
ax.set_ylabel("Sales", fontsize=14);
```

### 2. Salesof 2020 vs country

```python
# plotting the data above

fig, ax = plt.subplots(figsize=(10,8))

x = year_2020['country']
y = year_2020['sales']

ax = sns.barplot(x = x, y = y)
plt.xticks(rotation = 90)

ax.set_title('Sales of 2020 Vs Country', fontsize=16)
ax.set_xlabel("Country",fontsize=14)
ax.set_ylabel("Sales", fontsize=14);
```

### Sales of 2021 vs country

```python
# plotting the data above

fig, ax = plt.subplots(figsize=(10,8))

x = year_2021['country']
y = year_2021['sales']

ax = sns.barplot(x = x, y = y)
plt.xticks(rotation = 90)

ax.set_title('Sales of 2021 Vs Country', fontsize=16)
ax.set_xlabel("Country",fontsize=14)
ax.set_ylabel("Sales", fontsize=14);
```

### 3. Total sales per each year

```python
#sum of sales per year
total_2019 = sum(year_2019['sales'])
total_2020 = sum(year_2020['sales'])
total_2021 = sum(year_2021['sales'])

#creating a dataframe
data = {'year':[2019, 2020, 2021],'total sales':[total_2019, total_2020, total_2021]}

sales_df = pd.DataFrame(data)

sales_df
```

```python
# plotting the data above

fig, ax = plt.subplots(figsize=(6,4))

x = sales_df['year']
y = sales_df['total sales']

ax = sns.barplot(x = x, y = y)

ax.set_title('Total Sales Vs Year', fontsize=16)
ax.set_xlabel("Year",fontsize=14)
ax.set_ylabel("Sales", fontsize=14);
```
